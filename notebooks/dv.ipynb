```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("quotes.csv")
# Plotting the Polarity and Subjectivity Distribution by Author
plt.figure(figsize=(14, 7))

# Plotting Polarity by Author (Boxplot)
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Author', y='Polarity', palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Polarity Distribution by Author')
plt.xlabel('Author')
plt.ylabel('Polarity')

# Plotting Subjectivity by Author (Boxplot)
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Author', y='Subjectivity', palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Subjectivity Distribution by Author')
plt.xlabel('Author')
plt.ylabel('Subjectivity')

plt.tight_layout()
plt.show()
```
